## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(['EIN','NAME'], axis=1, inplace=True)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value (Chose 300) and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
app_type = application_df['APPLICATION_TYPE'].unique()
app_type_value_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = [x for x in app_type if app_type_value_counts[x]<300]
print('Apps replaced: ', application_types_to_replace)
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#need to store in a new variable for the next questions
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
gt_1_counts = classification_counts[classification_counts>1].index
classifications_gt_1 = application_df[application_df['CLASSIFICATION'].isin(gt_1_counts)]['CLASSIFICATION'].value_counts()
classifications_gt_1
```

```python
# Choose a cutoff value (chose 1250) and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_type = application_df['CLASSIFICATION'].unique()
class_type_value_counts = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = [x for x in class_type if class_type_value_counts[x] < 1250]
print('Classes replaced: ', classifications_to_replace)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
#Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_num = pd.get_dummies(application_df, columns=['AFFILIATION','USE_CASE','ORGANIZATION','INCOME_AMT','CLASSIFICATION',
                                                 'APPLICATION_TYPE','SPECIAL_CONSIDERATIONS'])
df_num.head()
```

```python
# Split our preprocessed data into our features and target arrays
x = df_num.drop('IS_SUCCESSFUL', axis=1).values
y = df_num['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1,stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

## Compile, Train and Evaluate the Model

```python
#Figure out how many features we have
print(x_train.shape[1])
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#We should have at least more neurons than inputs. Since we have 43 inputs, we should have more than 43 . Going with 1.5*43

nn = tf.keras.models.Sequential()
#Input Layer
nn.add(tf.keras.layers.Dense(units=86, activation='relu',input_dim=43))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=86, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=86, activation='relu'))

#Third hidden layer
nn.add(tf.keras.layers.Dense(units=86, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
nn.fit(x_train_scaled, y_train, epochs=150)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity3.h5')
```
